```python
import pandas as pd
from scipy.io import loadmat

data_dict = loadmat('data/cars_train_annos.mat')
data_array = data_dict['annotations']
data_array = data_array.transpose(1, 0)
df = pd.DataFrame(data_array,
                  columns=['ch'+str(n) for n in range(1,7)])
```

```python
bx1 = []
by1 = []
bx2 = []
by2 = []
cats = []
names = []
```

```python
for i in data_array:
    print(i)
    x1 = i[0][0][0][0]
    y1 = i[0][1][0][0]
    x2 = i[0][2][0][0]
    y2 = i[0][3][0][0]
    category = i[0][4][0][0]
    imgname = i[0][5][0]
    
    bx1.append(x1)
    by1.append(y1)
    bx2.append(x2)
    by2.append(y2)
    cats.append(category)
    names.append(imgname)
```

```python
data = {"img_name":names, "x1":bx1,"y1":by1,"x2":bx2,"y2":by2, "class": cats}
print(data)
```

```python
df = pd.DataFrame(data=data)
df.head()
```

```python
df.to_csv("data/data.csv")
```
